```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
%matplotlib
from  tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
```

```python
# делим наш датасат на обучачающую и тестовую выборку
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Cout', 'Sandal', 'Shirt', 'Sneaker','Bag', 'Ankle boot']
```

```python
# Предварительная обработка данных
```

```python
# Посмотрим как выглядит изображение
```

```python
plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.grid(False)
```

```python
# Нормализация данных
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.grid (False)
```

```python
# просмотрим несколько изображений
```

```python
plt.figure(figsize=(10,10))
for i in range (25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_train[i])
  plt.xlabel(class_names[y_train[i]])
```

```python
# создание нейронной сети
```

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(128, activation='relu'),
                          keras.layers.Dense(10, activation='softmax')
])
```

```python
# Компиляция модели
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(), loss='sparse_categorical_crossentroppy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Обучение модели
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```
